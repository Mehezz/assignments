** Step-3 Data Processing**

```
import nltk
nltk.download("stopwords")
nltk.download("wordnet")
```

```
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```
stemmer = PorterStemmer()
```

```
lemmatizer = WordNetLemmatizer()
```

```
new_clean_df_lst = []

df_len_lst = []
```

```
def preprocess(raw_df,flag):
 
  only_letters = re.sub("[^a-zA-Z]"," ",raw_df)
 
  only_letters = only_letters.lower()

  words = only_letters.split()
 
  words = [w for w in words if not w in stopwords.words("english")]
 
  if(flag== ' stem'):
    words = [stemmer.stem(word) for word in words]
  else:  
    words = [lemmatizer.lemmatize(word) for word in words]
 
    new_clean_df_lst.append(" ".join(words))
 
    df_len_lst.append(len(words))
```

```
 
from tqdm import tqdm, tqdm_notebook
 
tqdm.pandas()
```

```
new_clean_df_lst = []
 
df_len_lst=[]
 
df['message'].progress_apply(lambda x: preprocess(x, 'stem'))
 
df['new_clean_df_stem'] = new_clean_df_lst
 
df['df_length_stem'] = df_len_lst
```

```
new_clean_df_lst = []
 
df_len_lst=[]
 
df['message'].progress_apply(lambda x: preprocess(x, 'stem'))
 
df['new_clean_df_lema'] = new_clean_df_lst
 
df['df_length_lema'] = df_len_lst
```

```
df.head()
```

** Step-4 Data Preparation**

```
df['type'] = df['target'].apply(lambda x: 0 if x=='spam' else 1)

df.head()
```

```
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size = 0.25 , random_state=40)
```

```
train_new_clean_df = []

for df in train['new_clean_df_stem']:
  train_new_clean_df.append(df)

test_new_clean_df = []

for df in test['new_clean_df_stem']:
  test_new_clean_df.append(df)
```

```
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = 'word')

train_features = vectorizer.fit_transform(train_new_clean_df)

test_features = vectorizer.transform(test_new_clean_df)
```

```
vectorizer.vocabulary_
```

```
print('Number of unique word present = ', len(vectorizer.vocabulary_))

print('type of train_features = ', type(train_features))

print('type of test_features = ', type(test_features))

print('shape of input data = ', train_features.shape)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
 
from sklearn.metrics import accuracy_score
```

```
Classifiers = [LogisticRegression(), DecisionTreeClassifier(), SVC(kernel='linear')]
```

```
 x_features = train_features.toarray()
 
x_test = test_features.toarray()
 
for classifier in tqdm(Classifiers):
    fit = classifier.fit(x_features,train['type'])
    pred = fit.predict(x_test)
    accuracy = accuracy_score(pred,test['type'])
    print('Accuracy of '+classifier.__class__.__name__+' is '+str(accuracy))
```
